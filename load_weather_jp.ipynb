# Get all dates in the target period
From 2004 to 2016

```python
d1 = datetime(2004, 1,1)
d2 = datetime(2016,12,31)

delta = d2-d1
```

```python
span = []
for i in range(delta.days+1):
     intdate = map(int, str(d1+timedelta(days=i)).split(' ')[0].split('-'))
     span.append(intdate)
print len(span)
```

```python
frames =[]
for day in span:
    date = (day[0],day[1],day[2])
    print date
    url = 'http://www.data.jma.go.jp/obd/stats/etrn/view/10min_a1.php?prec_no=44&block_no=1002&year=%s&month=%s&day=%s&view=' %(date)
    d = pd.read_html(url)[0]
    d = d.ix[2:145]
    d['year'], d['month'], d['day']=date
    frames.append(d)

total = pd.concat(frames)
hd = ['時分', '降水量(mm)', '気温(℃)', '平均風速(m/s)','平均風向','最大瞬間風速(m/s)', '最大瞬間風向','日照時間(分)','year','month','day']
hd_utf = [i.decode('utf-8') for i in hd]
total.columns= hd_utf
total['location'] = '練馬'.decode('utf-8')

total.to_csv('total_weather.csv', index= False, header = True , encoding = 'utf-8')
```
